```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib tk
```

```python
folder="../data/third/"
```

```python
fig,ax=plt.subplots()
temperature=[]
max_power=[]
wavelength=[]
names=['Wavelength (nm)','Power (dBm)']
for file in os.listdir(folder):
    if file.endswith('csv'):
        file_name=os.path.join(folder,file)
        df=pd.read_csv(file_name,names=names,skiprows=28)
        temp=file.strip('data'+'.csv').replace('_','.')
        temperature.append(temp)
        mx=max(df['Power (dBm)'])
        lmx=df.loc[df['Power (dBm)'] == mx, 'Wavelength (nm)'].values[0] 
        max_power.append(mx)
        wavelength.append(lmx)
        ax.plot(df['Wavelength (nm)'],10*np.log10(df['Power (dBm)']),'k-',label=temp)
ax.set_xlim(1545,1555)
ax.set_xlabel(r'$\lambda (nm)$',fontsize=14)
ax.set_ylabel(r'Power (dBm)',fontsize=14)
ax.grid(True)
```

```python
fig,ax=plt.subplots()
#ax.plot(temperature,max_power)
ax.plot(temperature,wavelength,'r',label='$\lambda_{peak}$')
ax2=ax.twinx()
ax2.plot(temperature,max_power,'b',label='power')
ax.set_ylabel(r'$\lambda_{peak} (nm)$',fontsize=14)
ax2.set_ylabel(r'$Power (dBm)$',fontsize=14)
ax.set_xlabel(r'Temperature (K)',fontsize=14)
ax.grid(True)
ax.legend(loc='lower right')
ax2.legend(loc='lower left')
```

## Some data

```python
folder="../data/third/sub_data/"
temperature2=[]
max_power2=[]
wavelength2=[]
##
files = [file for file in os.listdir(folder)]
files_o = sorted(files, key=lambda x: os.path.getctime(os.path.join(folder, x)))


fig,ax=plt.subplots()
names=['Wavelength (nm)','Power (dBm)']
for file in files:
    a1,a2=200,300
    if file.endswith('csv') and file.startswith('data2'):
        file_name=os.path.join(folder,file)
        df=pd.read_csv(file_name,names=names,skiprows=28)
        temp=file.strip('data'+'.csv').replace('_','.')
        temperature2.append(temp)
        mx=max(df['Power (dBm)'][a1:a2])
        lmx=df.loc[df['Power (dBm)'] == mx, 'Wavelength (nm)'].values[0] 
        max_power2.append(10*np.log10(mx))
        wavelength2.append(lmx)
        ax.plot(df['Wavelength (nm)'][a1:a2],10*np.log10(df['Power (dBm)'][a1:a2]),'-*',label='T='+temp+'K')
        #ax.axvline(lmx,linestyle='--',color='g')
    a1,a2=0,50
    if file.endswith('csv') and file.startswith('data1'):
        file_name=os.path.join(folder,file)
        df=pd.read_csv(file_name,names=names,skiprows=28)
        temp=file.strip('data'+'.csv').replace('_','.')
        temperature2.append(temp)
        mx=max(df['Power (dBm)'][a1:a2])
        lmx=df.loc[df['Power (dBm)'] == mx, 'Wavelength (nm)'].values[0] 
        max_power2.append(10*np.log10(mx))
        wavelength2.append(lmx)
        ax.plot(df['Wavelength (nm)'][a1:a2],10*np.log10(df['Power (dBm)'][a1:a2]),'-*',label='T='+temp+'K')
        #ax.axvline(lmx,linestyle='--',color='g')

#ax.plot(df['Wavelength (nm)'],10*np.log10(df['Power (dBm)']),label=temp+'K')

ax.set_xlabel(r'$\lambda (nm)$',fontsize=14)
ax.set_ylabel(r'Power (dBm)',fontsize=14)
#ax.set_xlim(1545,1555)
ax.legend()
ax.grid(True)
```

```python
datam={'Temperature(K)': temperature2, 'Max_Power(dBm)': max_power2, 'lambda(nm)': wavelength2}
dfm=pd.DataFrame(datam).sort_values(by='Temperature(K)',ascending=False).astype(float)
lambda0,t0=1550.52,297.13
dfm['deltaT'] = t0-dfm['Temperature(K)']
dfm['deltaL'] = lambda0-dfm['lambda(nm)']
dfm
```

```python
fig,ax=plt.subplots()
ax.plot(dfm['Temperature(K)'],dfm['lambda(nm)'],'r-*',label='$\lambda_{peak}$')
ax2=ax.twinx()
ax2.plot(dfm['Temperature(K)'],dfm['Max_Power(dBm)'],'b-*',label='power')

ax.set_ylabel(r'$\lambda_{peak} (nm)$',fontsize=14)
ax2.set_ylabel(r'Max Power (dBm)',fontsize=14)
ax.set_xlabel(r'Temperature (K)',fontsize=14)

ax.grid(True)
ax.legend(loc='upper left')
ax2.legend(loc='lower right')
# ax2.legend(loc='lower left')
```

```python
fig,ax=plt.subplots()
x=dfm['deltaT'][1:]
y=dfm['deltaL'][1:]/lambda0
ax.plot(x,y,'*',label='data')

ax.set_ylabel(r'$(\lambda-\lambda_0)/\lambda_0$',fontsize=14)
ax.set_xlabel(r'$\Delta T$ (K)',fontsize=14)

ax.grid(True)
## Fitting
slope, intercept = np.polyfit(x, y, 1)

# Generar valores para la recta ajustada
x_fit = np.linspace(min(x), max(x), 100)
y_fit = slope * x_fit + intercept

# graph
plt.plot(x_fit, y_fit, 'r-', label='Fitting m={}'.format(slope))

ax.legend(loc='upper left')
```
