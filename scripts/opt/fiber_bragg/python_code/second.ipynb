```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
folder="../data/second/"
```

```python
fig,ax=plt.subplots()
temperature=[]
max_power=[]
wavelength=[]
names=['Wavelength (nm)','Power (dBm)']
for file in os.listdir(folder):
    if file.endswith('csv'):
        file_name=os.path.join(folder,file)
        df=pd.read_csv(file_name,names=names,skiprows=28)
        temp=file.strip('data'+'.csv').replace('_','.')
        temperature.append(temp)
        mx=max(df['Power (dBm)'])
        lmx=df.loc[df['Power (dBm)'] == mx, 'Wavelength (nm)'].values[0] 
        max_power.append(mx)
        wavelength.append(lmx)
        ax.plot(df['Wavelength (nm)'],10*np.log10(df['Power (dBm)']),label=temp)

ax.set_xlabel(r'$\lambda (nm)$',fontsize=14)
ax.set_ylabel(r'Power (dBm)',fontsize=14)
ax.grid(True)
```

```python
fig,ax=plt.subplots()
#ax.plot(temperature,max_power)
ax.plot(temperature,wavelength,'r',label='$\lambda_{peak}$')
ax2=ax.twinx()
ax2.plot(temperature,max_power,'b',label='power')
ax.set_ylabel(r'$\lambda_{peak} (nm)$',fontsize=14)
ax.set_xlabel(r'Temperature (K)',fontsize=14)
ax.grid(True)
ax.legend(loc='lower right')
ax2.legend(loc='lower left')
```
